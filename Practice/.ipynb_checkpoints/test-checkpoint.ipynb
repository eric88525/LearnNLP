```python
m = 2
r = 2

a = torch.ones([m,r,4])
a = a.unsqueeze(-2)
print(a)
print(a.shape)

b = torch.randn([m,r,4])
b = b.unsqueeze(-3)
print(b)
print(b.shape)

c = a+b
print(c)
print(c.shape)

d = c.narrow(-3,0,c.size(-3)-1)
print(d,d.shape)
```
